```python
airports_in_sources      = airports_df['Airport_ID'].isin(routes_df['Source_ID'])
airports_in_destinations = airports_df['Airport_ID'].isin(routes_df['Destination_ID'])


print('In Sources:')
print(len(airports_df[(airports_in_sources)]))
print('In Destinations:')
print(len(airports_df[(airports_in_destinations)]))
print('Not in Sources but in Routes:')
print(len(airports_df[(~airports_in_sources)]))
print('Not in Destinations but in Destinations:')
print(len(airports_df[(~airports_in_destinations)]))

#How many airports are not included in a route?
print('How many airports are not included in a route?')
q1a = len(airports_df[(~airports_in_sources) & (~airports_in_destinations)])
print(q1a)

#How many airports serve as both source and destination?
print('How many airports serve as both source and destination?')
q2a = len(airports_df[(airports_in_sources) & (airports_in_destinations)])
print(q2a)

print('Unutilized as Source / Destination')
print(q1a - q2a)
```

```python
# TODO: get random sample of flights from SQL
query = 'SELECT * FROM flights_df ORDER BY RANDOM() LIMIT 1000'
flights_s_df = pd.read_sql_query(query, con=engine)
```

```python
# TODO: create nested map of Jaccard distance

def get_jaccard(a, b):
    s1, s2 = set(a), set(b)
    return len(s1 & s2) / float(len(s1 | s2))

map_jaccard = {}
r_cols = ['Airline', 'Source_Airport','Destination_Airport']
f_cols = ['Carrier','Origin', 'Dest']
for col_r in routes_df[r_cols]:
    for col_f in flights_s_df[f_cols]:
        a, b = flights_s_df[col_f].drop_duplicates(), routes_df[col_r].drop_duplicates()
        r, f, dist_j = str(col_r), str(col_f), get_jaccard(a, b)
        if r not in map_jaccard:
            map_jaccard[r] = {}
        if dist_j != 0:
            map_jaccard[r].update({f : dist_j})
```
